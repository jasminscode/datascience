# Initiale Exploration des Datensatzes

Um den Datensatz initial zu erkunden und uns einen Überblick zu verschaffen, verwenden wir ydata profiling.

> Warnung: Das Generieren des Profils dauert einige Minute, da der Datensatz sehr groß ist und es frisst viel RAM. Falls der Kernel mal abstürzt, kann man alle Notebooks schließen und den Kernel neu starten um den RAM zu leeren.

```python
from kagglehub import dataset_load, KaggleDatasetAdapter
from ydata_profiling import ProfileReport
import gc

# Download the dataset from Kaggle in the specified version
df = dataset_load(
    KaggleDatasetAdapter.PANDAS,
    "asaniczka/tmdb-movies-dataset-2023-930k-movies/versions/518",
    "TMDB_movie_dataset_v11.csv",)

profile = ProfileReport(df, title="Movie Dataset Profiling Report")
profile.to_file("index.html")

del df
del profile
gc.collect()
```
